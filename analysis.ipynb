```python
import pandas as pd

loan_data = pd.read_csv("loan_data.csv")
loan_data.head()
```

```python
# Parse the date columns into datetime
loan_data['start_date'] = pd.to_datetime(loan_data['start_date'])
loan_data['end_date'] = pd.to_datetime(loan_data['end_date'])
loan_data['interest_rate'] = pd.to_numeric(loan_data['interest_rate'])
loan_data.head()
```

```python
loan_data['loan_amount'] = loan_data['loan_amount'].astype(float)
loan_data['current_balance'] = loan_data['current_balance'].astype(float)
```

```python
def total_outstanding_balance(df):
    return df['current_balance'].sum()

def avg_ir_by_sector(df):
    avg_by_sec = df.groupby('sector')['interest_rate'].mean()

def default_rate(df):
    return (df['status'] == 'Default') / len(df['status'])
```

```python
def add_risk_flag(df):
    df['balance_ratio'] = df['current_balance'] / df['loan_amount']

    df['risk_flag'] = df.apply(
        lambda row: "High" if (row['status'] == 'Late' or row['status'] == 'Default') or row['balance_ratio'] < 0.4 else
        'Low' if row['balance_ratio'] > 0.6 else
        'Medium',
        axis=1
    )

    return df
```
